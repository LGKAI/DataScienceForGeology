```python
df1 = df.groupby('Rock')
```

```python
BASALT = df1.get_group('BASALT')
RHYOLITE = df1.get_group('RHYOLITE')
KOMATIITE = df1.get_group('KOMATIITE')
ANDESITE = df1.get_group('ANDESITE')
THOLEIITE = df1.get_group('THOLEIITE')
GABBRO = df1.get_group('GABBRO')
DACITE = df1.get_group('DACITE')
```

```python
plt.scatter(
    df1.get_group('BASALT')['SiO2'],
    df1.get_group('BASALT')['FeOT'],
    edgecolors = 'black',
    color = 'red',
    marker = '^',
    s=40,
    label='BASALT')
plt.scatter(
    df1.get_group('RHYOLITE')['SiO2'],
    df1.get_group('RHYOLITE')['FeOT'],
    edgecolors = 'black',
    color = 'blue',
    marker = '*',
    s=40,
    label='RHYOLITE')
plt.scatter(
    df1.get_group('KOMATIITE')['SiO2'],
    df1.get_group('KOMATIITE')['FeOT'],
    edgecolors = 'black',
    color = 'yellow',
    marker = 's',
    s=40,
    label='KOMATIITE')
plt.scatter(
    df1.get_group('ANDESITE')['SiO2'],
    df1.get_group('ANDESITE')['FeOT'],
    edgecolors = 'black',
    color = 'green',
    marker = 'o',
    s=40,
    label='ANDESITE')
plt.scatter(
    df1.get_group('THOLEIITE')['SiO2'],
    df1.get_group('THOLEIITE')['FeOT'],
    edgecolors = 'black',
    color = 'purple',
    marker = 'd',
    s=40,
    label='THOLEIITE')
plt.scatter(
    df1.get_group('GABBRO')['SiO2'],
    df1.get_group('GABBRO')['FeOT'],
    edgecolors = 'black',
    color = 'pink',
    marker = 'x',
    s=40,
    label='GABBRO')
plt.scatter(
    df1.get_group('DACITE')['SiO2'],
    df1.get_group('DACITE')['FeOT'],
    edgecolors = 'black',
    color = 'orange',
    marker = 'H',
    s=40,
    label='DACITE')

plt.xlabel('SiO2 [%]')
plt.ylabel('FeOT [%]')
plt.title('Biểu đồ phân tán SiO2 [%] vs FeOT [%]')
plt.legend()
plt.show()
```

# ***Nhận xét: Về sự phân bố SiO₂ và FeOT của từng nhóm đá***

***-Basalt***

Tập trung ở vùng SiO₂ thấp (45–52%) và FeOT cao (10–22%).

Phản ánh bản chất mafic giàu Fe–Mg.

***-Komatiite***

Có SiO₂ rất cao (≈40–48%) và FeOT cao từ (~10-15%)

Phù hợp với đặc trưng ultramafic, giàu Fe và Mg.

***-Tholeiite***

Phân bố gần basalt nhưng có SiO₂ phân bố nhiều ở khoảng (≈45–55%) và FeOT ~(9-17%)

Thể hiện xu hướng giảm FeOT khi SiO₂ tăng — đặc trưng tiến hóa magma.

*-Andesite*

Tập trung ở SiO₂ trung bình (53–65%) và FeOT thấp hơn basalt(~5-13%).

Phù hợp bản chất trung tính, ít Fe hơn nhóm mafic.

***-Dacite***

Có SiO₂ cao (65–72%), FeOT thấp (≈2–8%).

Phù hợp với magma trung tính–felsic, tiến hóa mạnh.

***-Rhyolite***

Nằm ở vùng SiO₂ rất cao (75–90%) và FeOT rất thấp (≈0–5%).

Thể hiện mức độ phân dị cao nhất trong các nhóm đá.

***-Gabbro***

Phân bố tương tự basalt nhưng SiO₂ và FeOT thấp biến thiên hơn, phù hợp với bản chất plutonic đồng nguồn.

****Xu hướng tổng quát***

FeOT giảm dần khi SiO₂ tăng, thể hiện quy luật tiến hóa magma từ mafic → intermediate → felsic.

Các nhóm đá phân tách khá rõ ràng theo trường SiO₂–FeOT, phản ánh mức độ tiến hóa khác nhau của hệ magma.

***Rhyolite:***
Phân bố hẹp hơn basalt. Median nằm khoảng 12–13%. Vẫn xuất hiện một vài outlier cao đến ~21–22%.

***Komatiite:***
Đây là nhóm có Al₂O₃ thấp nhất. Median chỉ tầm 8–9%. Một số điểm thấp hơn nữa (khoảng ~3%). Phạm vi thay đổi nhỏ, đúng với đặc điểm đá ultramafic nghèo Al.

***Andesite:***
Các giá trị tương đối ổn định. Median khoảng 15–16%. Khoảng dao động không quá lớn, ít outlier.

***Tholeiite:***
Median khoảng 13–14%, thấp hơn andesite. Có vài outlier thấp đến ~2%, cho thấy một số mẫu có hàm lượng Al₂O₃ khá bất thường.

***Gabbro:***
Dải phân bố tương đối rộng nhưng ổn định. Median khoảng 15–16%, khá giống andesite.

***Dacite:***

Dacite, andesite và gabbro nằm trong vùng có Al₂O₃ cao và ổn định nhất.

Các nhóm đá có sự phân tách khá rõ về mức Al₂O₃, phù hợp với mức độ tiến hóa của từng loại đá.

#Câu 4: Vẽ biểu đồ violin plot cho biến SiO2 cho các nhóm trên cùng một biểu đồ, đưa ra nhận xét về dữ liệu và các nhóm.



```python
groups = [BASALT['SiO2'],RHYOLITE['SiO2'],KOMATIITE['SiO2'],ANDESITE['SiO2'],THOLEIITE['SiO2'],GABBRO['SiO2'],DACITE['SiO2']]
labels = ['BASALT','RHYOLITE','KOMATIITE','ANDESITE','THOLEIITE','GABBRO','DACITE']
plt.violinplot(groups, showextrema=False)
plt.boxplot(groups, widths=0.15, tick_labels=labels)
plt.xlabel("Rock")
plt.ylabel("SiO2 (%)")
plt.title("Phân bố SiO2 theo từng nhóm Rock")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

***Nhận xét về dữ liệu và các nhóm:***
Nhìn vào biểu đồ violin, mỗi nhóm đá có một khoảng SiO₂ khá đặc trưng, và sự khác biệt giữa các nhóm khá rõ.

***Basalt:*** SiO₂ tập trung chủ yếu quanh 50%, nhưng dao động khá rộng từ khoảng 35–65%. Có vài giá trị thấp tới ~25% và cao tới ~88%, cho thấy nhóm này biến thiên mạnh hơn mong đợi.

***Rhyolite:*** Đây là nhóm giàu SiO₂ nhất, tập trung rõ rệt ở 70–80%. Dải dao động hẹp hơn, chủ yếu trong 65–85%, nên nhìn vào là biết ngay mẫu thuộc nhóm felsic.

***Komatiite:*** Nhóm này nghèo SiO₂, chủ yếu quanh 45–50%. Dải dao động nhỏ, nằm trong khoảng ~38–55%, thể hiện tính chất siêu mafic.

***Andesite:*** Có SiO₂ trung bình, tập trung khoảng 55–60%. Khoảng phân bố của nhóm này nằm trong vùng 45–70%, phản ánh đặc điểm trung gian giữa basalt và rhyolite.

***Tholeiite:*** Phân bố SiO₂ khá giống basalt nhưng tập trung thấp hơn một chút, khoảng 48–52%, với dải biến thiên khoảng 40–60%.

***Gabbro:*** Nhóm này SiO₂ thấp hơn andesite và basalt, chủ yếu quanh 45–50%, dao động trong vùng ~38–55%, đặc trưng của đá mafic.

***Dacite:*** Đây là nhóm SiO₂ cao thứ hai sau rhyolite, tập trung khoảng 65–70%. Dải dao động chủ yếu trong 58–75%, phù hợp với đặc điểm đá trung–felsic.

Tóm lại: Các nhóm đá giàu SiO₂ nhất: Rhyolite > Dacite.

Nhóm nghèo SiO₂ nhất: Komatiite, Gabbro.

Nhóm trung gian: Basalt, Tholeiite, Andesite.

Biểu đồ violin giúp thấy rõ vừa vị trí trung bình, vừa mức độ tản mát, và nhìn chung dữ liệu phù hợp với đặc điểm địa hoá của từng loại đá.

#Câu 5: Tính giá trị hệ số tương quan của dữ liệu và vẽ biểu đồ tương quan giữa các biến theo phương pháp Pearson.

```python
df_numeric = df.select_dtypes(include=np.number)
correlation_matrix = df_numeric.corr(method='pearson')
print("Hệ số tương quan (Pearson):")
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    linewidths=0.5,
    linecolor='black'
)
plt.title("Biểu đồ tương quan giữa các biến theo phương pháp Pearson.")
plt.show()
```

#Câu 6: Từ biểu đồ tương quan phía trên, từ đó vẽ biểu đồ mật độ hai chiều giữa 2 cột dữ liệu có độ tương quan cao nhất và biểu đồ mật độ hai chiều giữa 2 cột dữ liệu có độ tương quan thấp nhất. Nhận xét

```python
X1 = 'MgO'
Y1 = 'Al2O3'
plt.figure(figsize=(10, 8))
sns.kdeplot(
    data=df,
    x=X1,
    y=Y1,
    fill=True,
    cmap='Blues',
    levels=10
)
sns.scatterplot(data=df, x=X1, y=Y1, s=10, color='red', alpha=0.5)

plt.title(f'Mật độ 2D giữa {Y1} và {X1} (Tương quan cao)')
plt.xlabel(f'{X1} (%)')
plt.ylabel(f'{Y1} (%)')
plt.show()
```

```python
X2 = 'SiO2'
Y2 = 'Na2O'

plt.figure(figsize=(8, 7))
sns.kdeplot(
    data=df,
    x=X2,
    y=Y2,
    fill=True,
    cmap='Blues',
    levels=10
)
sns.scatterplot(data=df, x=X2, y=Y2, s=10, color='black', alpha=0.5)

plt.title(f'Mật độ 2D giữa {Y2} và {X2} (Tương quan thấp)')
plt.xlabel(f'{X2} (%)')
plt.ylabel(f'{Y2} (%)')
plt.show()
```

# ***Nhận xét biểu đồ tương quan cao mật đồ 2 chiều giữa Al2O3 và MgO***
Ở biểu đồ Al₂O₃–MgO, mật độ điểm tập trung rất rõ ở vùng MgO từ 0–10% và Al₂O₃ khoảng 12–18%. Vùng này tạo thành một “đám mây” đậm màu, cho thấy hai oxit này thay đổi khá nhất quán theo nhau. Khi MgO tăng lên trên 15–30%, thì Al₂O₃ giảm dần còn khoảng 5–10%, tạo thành xu hướng giảm khá rõ rệt. Một số điểm rải rác tới MgO ~45–50%, nhưng mật độ rất thấp. Nhìn chung, dữ liệu thể hiện mối quan hệ nghịch khá mạnh: đá càng giàu MgO thì lượng Al₂O₃ càng giảm.

# ***Nhận xét biểu đồ tương quan thấp mật đồ 2 chiều giữa Na2O và SiO2***
Biểu đồ Na₂O–SiO₂ thì không thấy sự thay đổi rõ ràng giữa hai thành phần. Mật độ chủ yếu nằm trong khoảng SiO₂ 45–55% và Na₂O 1–3%, nhưng hướng lan toả của điểm khá “mở”, không tạo xu hướng tuyến tính hay nghịch như cặp trên. Khi SiO₂ tăng lên tới 70–80%, Na₂O vẫn dao động lỏng lẻo từ 0–6% mà không theo quy luật rõ rệt. Điều này phản ánh tương quan rất yếu, tức là mức SiO₂ tăng hay giảm cũng không ảnh hưởng nhiều đến Na₂O.
