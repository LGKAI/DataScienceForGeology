```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import pdist, squareform
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
```

# Bài 2

a. A(10, 5); B(1, 4); C(5, 8); D(9, 2); E(12, 10); F(15, 8); G(7, 7)

```python
X_a = np.array([
    [10, 5],  # A
    [1, 4],   # B
    [5, 8],   # C
    [9, 2],   # D
    [12, 10], # E
    [15, 8],  # F
    [7, 7]    # G
])
labels_a = ['A', 'B', 'C', 'D', 'E', 'F', 'G']

dist_matrix_a = squareform(pdist(X_a, metric='euclidean'))
df_dist_a = pd.DataFrame(dist_matrix_a, columns=labels_a, index=labels_a)
print("Ma trận khoảng cách:")
print(df_dist_a.round(2))

print("\nDendrogram:")
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
methods = ['single', 'complete', 'average']
titles = ['Single Linkage', 'Complete Linkage', 'Average Linkage']
for i, method in enumerate(methods):
    Z = linkage(X_a, method=method, metric='euclidean')
    dendrogram(Z, labels=labels_a, ax=axes[i])
    axes[i].set_title(titles[i])
    axes[i].set_xlabel('Điểm dữ liệu')
    axes[i].set_ylabel('Khoảng cách')
    axes[i].grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# Bài 3

a. Hãy đọc và in ra 5 dòng đầu tiên của dữ liệu. Và kiểm tra thông tin của dữ liệu, đồng thời in ra các thống kê mô tả cho từng nhóm dữ liệu.

```python
df = pd.read_csv('geochem.csv')
print(df.head(5))
df.info()
print(df.describe())
```

b. Hãy vẽ biểu đồ khuỷu tay (scree plot) cho bộ dữ liệu, từ đó xác định số lượng thành phần chính cần được giữ lại cho phương
pháp giảm số chiều. Giải thích số thành phần chính mà các bạn lựa chọn.

```python
df_numeric = df.select_dtypes(include=[np.number])
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_numeric)
pca = PCA()
pca.fit(X_scaled)

explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
components = range(1, len(explained_variance) + 1)

plt.figure(figsize=(10, 6))
plt.plot(components, cumulative_variance, marker='o', linestyle='-', color='blue', label='Phương sai tích lũy')
plt.bar(components, explained_variance, alpha=0.6, color='orange', label='Phương sai từng thành phần')
plt.axhline(y=0.8, color='red', linestyle='--', label='Ngưỡng 80%')

plt.title('Biểu đồ Scree Plot (Khuỷu tay)')
plt.xlabel('Số lượng thành phần chính (Principal Components)')
plt.ylabel('Tỷ lệ phương sai giải thích')
plt.xticks(components)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

Quan sát biểu đồ Scree Plot, tôi quyết định giữ lại 2 thành phần chính (hoặc 3 tùy vào hình thực tế). Lý do là tại PC2 (hoặc PC3), tổng phương sai tích lũy đã đạt trên 80% lượng thông tin của dữ liệu gốc. Đồng thời, đây cũng là điểm 'khuỷu tay' của đồ thị, khi sự đóng góp của các thành phần phía sau (PC3, PC4...) giảm đi rõ rệt.

c. Áp dụng phương pháp phân cụm theo thứ bậc với phương pháp 'complete' với 2 cột dữ liệu 'K2O', 'Na2O'. Vẽ biểu đồ
Dendrogram và biểu đồ phân tán theo cụm. Giải thích về việc lựa chọn số cụm, và nhận xét biểu đồ.

```python
X_c = df[['K2O', 'Na2O']]
plt.figure(figsize=(10, 5))
plt.title("Dendrogram (K2O, Na2O) - Phương pháp Complete")
dend = shc.dendrogram(shc.linkage(X_c, method='complete', metric='euclidean'))
plt.axhline(y=4, color='r', linestyle='--')
plt.show()

model_c = AgglomerativeClustering(n_clusters=3, linkage='complete', metric='euclidean')
df['Cluster_3c'] = model_c.fit_predict(X_c)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='K2O', y='Na2O', hue='Cluster_3c', palette='viridis', s=60)
plt.title('Phân cụm phân cấp (K2O vs Na2O)')
plt.show()
```

- Lựa chọn số cụm ($k$): Nhìn vào Dendrogram, bạn tìm khoảng cách trục tung lớn nhất mà không bị cắt bởi các đường ngang (khoảng cách giữa các lần gộp cụm). Kẻ một đường ngang qua khoảng đó, số đường dọc bị cắt chính là số cụm tối ưu (ví dụ trong code là 3).

- Nhận xét: Các điểm dữ liệu được gom thành các nhóm dựa trên hàm lượng kiềm (K2O và Na2O). Các nhóm này có thể đại diện cho các loại đá có tính kiềm thấp, trung bình hoặc cao khác nhau.

d. Áp dụng phương pháp phân cụm theo thứ bậc với phương pháp 'complete' với 3 cột dữ liệu 'K2O', 'Na2O', 'P2O5'. Vẽ biểu đồ
Dendrogram và biểu đồ phân tán theo cụm. Giải thích về việc lựa chọn số cụm, và nhận xét biểu đồ.

```python
X_d = df[['K2O', 'Na2O', 'P2O5']]
plt.figure(figsize=(10, 5))
plt.title("Dendrogram (3 biến) - Phương pháp Complete")
dend = shc.dendrogram(shc.linkage(X_d, method='complete', metric='euclidean'))
plt.show()

model_d = AgglomerativeClustering(n_clusters=3, linkage='complete', metric='euclidean')
df['Cluster_3d'] = model_d.fit_predict(X_d)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(df['K2O'], df['Na2O'], df['P2O5'], c=df['Cluster_3d'], cmap='viridis', s=50)
ax.set_xlabel('K2O')
ax.set_ylabel('Na2O')
ax.set_zlabel('P2O5')
plt.title('Phân cụm phân cấp 3D')
plt.colorbar(sc, label='Cluster')
plt.show()
```

- Lựa chọn số cụm: Tương tự câu 3c, dựa vào khoảng cách lớn nhất giữa các lần gộp trên Dendrogram.

- Nhận xét: Việc thêm biến P2O5 (Photpho) giúp tách biệt rõ hơn các mẫu đá mà ở không gian 2 chiều (K2O, Na2O) có thể bị chồng lấn. Biểu đồ 3D cho thấy sự phân bố không gian của các nhóm đá dựa trên 3 thành phần hóa học này.

e. Áp dụng phương pháp phân tích thành phần chính cho bộ dữ liệu theo số thành phần chính đã xác định ở Câu 3b. Từ dữ liệu
sau khi giữ lại số thành phần chính, hãy dùng K-Means để xác định số cụm và vẽ biplot kết hợp với các điểm dữ liệu được tô màu theo cụm, và các loading vector. Nhận xét các biểu đồ.

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

inertias = []
K_range = range(1, 10)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_pca)
    inertias.append(kmeans.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(K_range, inertias, 'bo-')
plt.title('Phương pháp Elbow tìm k tối ưu')
plt.xlabel('Số cụm k')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

k_opt = 3
kmeans = KMeans(n_clusters=k_opt, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_pca)

plt.figure(figsize=(10, 8))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=clusters, palette='deep', s=60, edgecolor='black')
loadings = pca.components_.T * 4

all_numeric_cols = df.select_dtypes(include=[np.number]).columns
features = [col for col in all_numeric_cols if 'Cluster' not in col]
features = features[:loadings.shape[0]]

for i, feature in enumerate(features):
    plt.arrow(0, 0, loadings[i, 0], loadings[i, 1], color='r', alpha=0.5, head_width=0.1)
    plt.text(loadings[i, 0]*1.15, loadings[i, 1]*1.15, feature, color='red', ha='center', va='center', weight='bold')

plt.title('Biplot: K-Means trên dữ liệu PCA')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

- Nhận xét biểu đồ Elbow: Chọn số cụm $k$ tại điểm "khuỷu tay" mà sau đó độ giảm của quán tính (inertia) không còn đáng kể (thường là 3).

- Nhận xét Biplot:
  + Các điểm dữ liệu cùng màu thể hiện các mẫu đá có đặc tính hóa học tương đồng nhau sau khi giảm chiều.

  + Các mũi tên đỏ (Loading vectors) cho biết hướng biến thiên của các oxit. Ví dụ: nếu vector $SiO_2$ và $MgO$ ngược chiều nhau, nghĩa là mẫu nào nhiều $SiO_2$ thì thường ít $MgO$ và ngược lại.

  + Các vector dài và gần trục PC1 hoặc PC2 nhất là những thành phần đóng góp chính vào sự phân chia các cụm này.
