d. Tạo và tính toán trên mảng data

```python
data = np.array([3, 7, 5, 10, 12, 8])
```

Tính giá trị trung bình

```python
# Cách 1: Dùng công thức
mean_manual = sum(data) / len(data)
print("Giá trị trung bình (dùng công thức):", mean_manual)

# Cách 2: Dùng NumPy
mean_numpy = np.mean(data)
print("Giá trị trung bình (dùng NumPy):", mean_numpy)
```

Tính phương sai và độ lệch chuẩn

```python
# Cách 1: Dùng công thức
squared_diff = [(x - mean_manual)**2 for x in data]
variance_manual = sum(squared_diff) / (len(data) - 1)
std_manual = math.sqrt(variance_manual)
print("Phương sai (dùng công thức):", variance_manual)
print("Độ lệch chuẩn (dùng công thức):", std_manual)

# Cách 2: Dùng NumPy
variance_np = np.var(data, ddof=1)
std_np = np.std(data, ddof=1)
print("Phương sai (dùng NumPy):", variance_np)
print("Độ lệch chuẩn (dùng NumPy):", std_np)
```

# Bài 2

Đọc file và in 5 dòng đầu tiên

```python
df = pd.read_csv("meuse.csv")
df.head(n = 5)
```

Trích ra thông tin các cột 'cadmium', 'lead', 'zinc'

```python
cols = df[['cadmium', 'lead', 'zinc']]
print(cols)
```

Tính giá trị trung bình và độ lệch chuẩn của cột 'zinc'

```python
zinc_mean = df['zinc'].mean()
zinc_std = df['zinc'].std(ddof=1)
print("Giá trị trung bình:", zinc_mean)
print("Độ lệch chuẩn:", zinc_std)
```

Tính các phân vị và khoảng tứ phân vị của cột 'cadmium'

```python
Q1 = df['cadmium'].quantile(0.25)
Q2 = df['cadmium'].quantile(0.5)
Q3 = df['cadmium'].quantile(0.75)
IQR = Q3 - Q1
print("Phân vị thứ nhất (Q1):", Q1)
print("Phân vị thứ hai (Q2 - median):", Q2)
print("Phân vị thứ ba (Q3):", Q3)
print("Khoảng tứ phân vị (IQR = Q3 - Q1):", IQR)
```
